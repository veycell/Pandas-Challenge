```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#number of total schools
TotalSchool=school_data_complete["school_name"].count()
#number of students
TotalStudents=school_data_complete["student_name"].count()
#total budget
TotalBudget=school_data_complete["budget"].unique().sum()
#mean math score
AverageMath=school_data_complete["math_score"].mean()
#mean reading score
AverageReading=school_data_complete["reading_score"].mean()
#Overall pass
OverallPass=(AverageReading+AverageMath)/2
#Pass Math
PassMath=school_data_complete[school_data_complete.math_score>=70]['math_score'].count()/TotalStudents*100
#Pass Reading
PassReading=school_data_complete[school_data_complete.reading_score>=70]['reading_score'].count()/TotalStudents*100

df=pd.DataFrame({"Total Schools":[TotalSchool],"Total Students":[TotalStudents],"Total Bugdets":[TotalBudget],
    "Average Math Score": [AverageMath],"Average Reading Score":[AverageReading],"% Passing Math":[PassMath],
    "% Passing Reading":[PassReading],"% Overall Passing Rate":[OverallPass]})
df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
SchoolStats=school_data_complete.groupby(["school_name"])
SchoolNames=SchoolStats["school_name"]
StudentsBySchools=SchoolStats["student_name"].count()
BudgetBySchool=SchoolStats["budget"].sum()/StudentsBySchools
BudgetByStudent=BudgetBySchool/StudentsBySchools
AverageMathScore=SchoolStats["math_score"].mean()
AverageReadingScore=SchoolStats["reading_score"].mean()
PassingMath=(school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()/StudentsBySchools)*100
PassingReading=(school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()/StudentsBySchools)*100
OverallPassingRate=(PassingMath+PassingReading)/2
SchoolSummary=pd.DataFrame({"School Type":SchoolNames,"Total Students":StudentsBySchools,"Total School Budget":BudgetBySchool
                           ,"Per Student Budget":BudgetByStudent,"Average Math Score":AverageMathScore,
                           "Average Reading Score":AverageReadingScore,"% Passing Math":PassingMath,
                            "% Passing Reading":PassingReading,"% Overall Passing Rate":OverallPassingRate})
SchoolSummary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
TopPerformingSchool=SchoolSummary.sort_values(by="% Overall Passing Rate",ascending=False)
TopPerformingSchool.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
BottomPerformingSchool=SchoolSummary.sort_values(by="% Overall Passing Rate",ascending=True)
BottomPerformingSchool.head(5)
```

## Scores by School Spending

  * Overall Passing Rate (Average of the above two)
